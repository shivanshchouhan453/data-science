```python
import matplotlib.pyplot as plt
x=['A','N','E']
y=[1,2,3]
plt.bar(x,y,width=0.5 ,edgecolor='black')
plt.title("simple bar plot", fontsize=19)
#  This argument adds a label to the bars. It can be a single label string or a list of label strings.
plt.xlabel("character value of graph", fontsize=25,color='r')
plt.ylabel("character value of graph", fontsize=25)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [4, 5, 6]
#The color of the edges of the bars. Defaults to 'black'.
plt.bar(x, y, width=0.5, color=['red', 'green', 'blue'], edgecolor='black', linewidth=2)
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.title("Bar Plot with Custom Parameters")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['A','N','E']
y=[1,2,3]
# draw the horizantail  bar plot in using function "barh"
plt.barh(x,y)
plt.title("simple bar plot", fontsize=19)
plt.xlabel("character value of graph", fontsize=25)
plt.ylabel("character value of graph", fontsize=25)
# ticks function use in rotat in word
plt.xticks(rotation=15)
plt.yticks(rotation=15)
plt.show()
#  Returns the current figure object.
plt.gcf()
```
